```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH = "Poster_Training/"
df_1 = pd.read_csv(PATH + "DQN_Cartpole_rewards_run1.csv", skipinitialspace=True)
df_2 = pd.read_csv(PATH + "DQN_Cartpole_rewards_run2.csv", skipinitialspace=True)
df_3 = pd.read_csv(PATH + "DQN_Cartpole_rewards_run3.csv", skipinitialspace=True)
display(df_1.head())
```

```python
fig, ax = plt.subplots()

df_1.plot(x='Episode', y='avg_return', label='Seed 1', ax=ax)
df_2.plot(x='Episode', y='avg_return', label='Seed 2', ax=ax)
df_3.plot(x='Episode', y='avg_return', label='Seed 3', ax=ax)
plt.title('DQN Cartpole Training')
plt.xlabel('Episode')
plt.ylabel('Average Return')
plt.show()
```

```python
PATH = "Poster_Training/"
df_1 = pd.read_csv(PATH + "DQN_Pendulum_rewards_run1.csv", skipinitialspace=True)
df_2 = pd.read_csv(PATH + "DQN_Pendulum_rewards_run2.csv", skipinitialspace=True)
df_3 = pd.read_csv(PATH + "DQN_Pendulum_rewards_run3.csv", skipinitialspace=True)
display(df_1.head())
```

```python
fig, ax = plt.subplots()

df_1.plot(x='Episode', y='avg_return', label='Seed 1', ax=ax)
df_2.plot(x='Episode', y='avg_return', label='Seed 2', ax=ax)
df_3.plot(x='Episode', y='avg_return', label='Seed 3', ax=ax)
plt.title('DQN Pendulum Training [Discretized to 31 Inputs]')
plt.xlabel('Episode')
plt.ylabel('Average Return')
plt.show()
```

```python

PATH = "Poster_Training/"
df_1 = pd.read_csv(PATH + "DQN_Mountaincar_rewards_run1.csv", skipinitialspace=True)

fig, ax = plt.subplots()

df_1.plot(x='Episode', y='avg_return', label='Seed 1', ax=ax)
ax.fill_between(df_1['Episode'], df_1['min_return'], df_1['max_return'], alpha = 0.1)

plt.title('DQN Mountaincar Training')
plt.xlabel('Episode')
plt.ylabel('Average Return')
plt.show()

```
